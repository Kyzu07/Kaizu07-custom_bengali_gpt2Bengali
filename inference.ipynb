```python
import torch
device = "cuda"
```

```python
max_length = 512
generation_kwargs = {
    "max_length": max_length,
    "num_return_sequences": 1,
    "temperature": 0.8,  # Example additional parameter
    "top_k": 50,
    "top_p": 0.90,
    "do_sample": True,
    "num_return_sequences": 2
}
```

```python
from transformers import pipeline, GPT2LMHeadModel, GPT2Tokenizer, AutoTokenizer, AutoModelForCausalLM
# import gradio as gr

model_path = "/kaggle/input/bn-lit-model/custom_bengali_gpt2Bengali"
model = GPT2LMHeadModel.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained("flax-community/gpt2-bengali")

# input_ids = tokenizer.encode(input_text, return_tensors="pt")

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, **generation_kwargs)

pipe('শীতকালে চাঁদপুরে বন্ধুদের সাথে ')
```

```python
import gradio as gr
demo = gr.Interface.from_pipeline(pipe)
demo.launch(debug=True)
```

```python
# Input text for completion
input_text = "কোথাও সে নাই, কেবল "
# Tokenize the input text
input_ids = tokenizer.encode(input_text, return_tensors="pt")
# Generate text completions
max_length = 512
output_text = model.generate(input_ids, max_length=max_length, top_k=50, top_p=.90, do_sample=True, num_return_sequences=2)[0]
# Decode the generated token IDs to text
completed_text = tokenizer.decode(output_text, skip_special_tokens=True)

print("Input Text:", input_text)
# print("Completed Text:", completed_text)
for line in completed_text.splitlines():
    print(line)
```
